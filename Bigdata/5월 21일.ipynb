### 📌 3. 불균형 샘플링 (106 Page)

샘플링 종류  (그림)
- 무작위(Random): 아무 규칙 없이 무심코 뽑기 (모든 데이터의 선택 확률이 동일).  
- 층별화(Stratified): 집단을 성격별(A, B, C 그룹)로 나누고, 각 그룹의 비율에 맞춰서 뽑기.  
- 계통(Systematic): 첫 번째를 정한 뒤 10번째, 20번째처럼 일정한 간격($k$)으로 뽑기.  
- 서브그룹: 특정 시간대나 구간별로 덩어리(단위)를 정해서 뽑기.  

##### 불균형 데이터 : 분류형 목표변수의 데이터가 불균형을 이룬 경우를 의미
##### 해결 : Over(소 -> 다 부풀) / Under(다 -> 줄여) / Hybird (O & U) Sampling

##### 불균형 데이터의 해결 방법 : Over Sampling : 적은 쪽 늘리기
1. S(synthehic) MOTE : 두 포인트를 연결하는 segment 상에 존재하는 data를 random sampling : 적은 데이터들 사이에 선을 긋고, 그 선 위에 가상의 데이터를 생성합니다.
2. BL(Borderline)SMOTE : borderline data 에 대하여 smote 적용 : 경계선에 있는 아슬아슬한 데이터들 근처에서 집중적으로 생성합니다.
3. DB(DBSCAN)SMOTE : DBSCAN cluster 생성 후, cluster내에서 smote 적용 : 비슷한 데이터끼리 뭉친 클러스터를 만든 후, 그 안에서 생성합니다.

##### 불균형 데이터 해결 방법 : Under Sampling : 많은 쪽 늘리기
1. Random Under Sampling : 무작위 삭제
2. Tomek links : 경계선 근처에서 서로 다른 종류끼리 너무 붙어 있는 경우, 많은 쪽의 데이터를 지워 경곌 깔끔하게 만듭니다.
3. EasyEnsemble/Balance Cascade : 데이터를 여러 번 쪼개서 샘플링 하고 학습시키는 과정을 반복하여 성능 높여

### 📌 이상치 이해 및 확인 (114 Page) : 그림 보세요 115, 116도요

이상치 이해 : 데이터 값이 발생할 수 있는 범위를 벗어난 데이터를 의미(프로세스 불안정, 급격한 변동 등)
- 이상치는 분석 모델에 영향을 많이 주고, 분석 결과를 왜곡시킬 수 있으므로 적절한 사전 처리가 필요하다.
- 이상치에 대한 적절한 처리는 분석 결과에대한 신뢰성 확보가 가능함.
- 이상치라고 해서 무조건 의미 없고 잘목된 값이라는 의미는 아니므로 충분한 사전 검토가 필요하다.
- 일반적인 이상치 판단 기준: 평균 +- 3 * 표준편자, Q3/Q1 (3, 1사분위) =- 1.5 * IQR(사분위 범위)를 벗어난 경우

- 변수 삾의 형태(2차식, log 함수 등)를 고려하여 파생변수를 생성하여 변환된 값으로 변경

#### 변수 변환 방법
#### One - Hot Encoding 
- 자료의 고유값 종류 만큼 변수 생성하고 각 자료가 고유값에 해당하면 1, else 0 : 이진화 방법
- 통계 또는 기계학습 분석에서 자주 활용되는 방법
- 값이 순서가 없거나(국가명) 값의 개수가 많지 않을때 효율적
#### Label - Encoding
- 자료 값을 정렬하여 값의 순서순으로 숫자 할당
- 실제 자료 값은 순서가 없으나 인코딩 하면 순서 생김 : 해석에 주의해야함
- 값의 순서가 의미 있을때 (직급-사원, 대리-과장) 고유값의 개수가 많은 경우

#### 크기 조정 121 Page
- standard scaler : 수능 표준점수 : 평균 - 표준편차를 이요한 변환 (평균 = 0, 표준편차 = 1) : 이상치 영향이 큼   
- : 평균이 0으로 가요
- min-max scaler: 최대 최소값을 이용한 변환 (최소값 = 0, 최대값 = 1) : 이상치 영향이 큼 
- : 최소값이 0으로 가요
- Robust scaler: 분위수를 이용한 변환 (중앙값 = 0, IQR = 1)

```python
#결측치를 문자로 바꾸자 fillna(값, inplace: True: 즉시 교체) / fillna(값, inplace: False: 원본을 바꾸지 않고 결과물(복사본)만 보여)
df_raw["GENDER"].fillna("여성", inplace=True)
df_raw.head()


# inplace=False는 원본을 바꾸지 않고 결과물(복사본)만 보여주며, 이는 pandas 함수의 기본 설정값입니다.
```

```python
# 숫자형 결측치 처리 :groupby(범주형 변수), agg("집계함수")
# ()별 []변수의 .agg(mean) 이면 평균
df_raw.groupby("GENDER")["WEIGHT"].agg("mean") 


#성별(GENDER)로 데이터를 그룹화한 뒤, 각 그룹에 해당하는 몸무게(WEIGHT)의 평균을 산출하는 집계 코드입니다
#groupby로 기준을 세우고 agg를 통해 특정 통계량(mean)을 적용하며, 결과는 성별별로 요약된 데이터 형태로 반환
```

```python
df_raw["WEIGHT"] = df_raw["WEIGHT"].fillna(df_raw.groupby("GENDER")["WEIGHT"].transform("mean")).round(3)
df_raw.head()
```

```python
# 'RUNTIME' 열의 문자열을 공백 기준으로 쪼개 숫자와 단위를 각각 'RUNTIME'과 'RUNTIME_UNIT' 열에 분리하여 저장합니다.
df_raw[['RUNTIME', 'RUNTIME_UNIT']] = df_raw['RUNTIME'].str.split(' ', expand = True)

# 문자열로 저장된 'RUNTIME' 열의 데이터를 산술 연산이 가능하도록 실수형(float64) 타입으로 변환합니다.
df_raw['RUNTIME'] = df_raw['RUNTIME'].astype('float64')
```

```python
# 수정 코드: astype(str)로 타입을 맞추고, 결과 중 앞의 2개 열만 선택(.iloc[:, :2])
df_raw[['RUNTIME', 'RUNTIME_UNIT']] = df_raw['RUNTIME'].astype(str).str.split(' ', expand=True).iloc[:, :2]
df_raw['RUNTIME'] = df_raw['RUNTIME'].astype('float64')
```

```python
# 137 Page
# 일자 변환 및 파생변수 생성
df_fitness['BIRTH'] = pd.to_datetime(df_fitness['BIRTH'], format = '%Y-%m-%d')
df_fitness['YEAR'] = df_fitness['BIRTH'].dt.year
df_fitness['MONTH'] = df_fitness['BIRTH'].dt.month
df_fitness['DAY'] = df_fitness['BIRTH'].dt.day
df_fitness['WEEKDAY'] = df_fitness['BIRTH'].dt.weekday
df_fitness.head()
```

```python
#데이터프레임에서 분석에 불필요한 특정 행(axis=0)이나 열(axis=1)을 지정하여 제거하는 함수입니다.
#df.drop(['삭제할열'], axis=1)과 같이 리스트 형태로 여러 열을 한꺼번에 지울 수 있으며, 보통 원본에 반영하기 위해 다시 변수에 저장합니다.
df_fitness = df_fitness.drop(['NAME', 'BIRTH', 'RUNTIME_UNIT'], axis=1)
df_fitness.head()
```
