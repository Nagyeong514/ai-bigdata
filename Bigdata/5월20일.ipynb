### 데이터 선택

- mask : 조건문이 참(True)인 데이터 포인트들을 골라내어 지정한 값으로 변경하거나 가려버리는(결측치 처리 등) 데이터 변형 함수입니다.

```python
# 37 Page (데이터선택 연습)
df = pd.DataFrame({'ABC': ['A','B','C','A'], 
                   'abc':['a','b','c','c'],
                   'group1':['G1','G2','G3','G4'],
                   'value1':[11,13,12,14],
                   'val1_mod':[0.231,0.197,0.213, 0.199] })
                   
df
```

axis=0: "이름(번호)이 1인 사람(행) 다 나와!" $\rightarrow$ 특정 데이터 한 줄 추출

### 데이터 결합 (MERGE)

DataFrame의 공통 열(column) 혹은 인덱스(index)를 기준으로 두 개의 DataFrame을 합침(변수 증가).
이 때 기준이 되는 열, 행을 키(key)라고 함


```python
# 55 Page (데이터결합 연습)
# left, right 데이터 생성
left = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
# merge 결과 result 생성
# 옵션 없음 : 기본값 : inner (교집합)
result = pd.merge(left, right, on=['key1','key2'])
result
```

```python
# 56 Page
# how = left
# left outer join
result = pd.merge(left, right, on=['key1','key2'], how='left')
result
```

```python
# how = right
# right outer join
result = pd.merge(left, right, on=['key1','key2'], how='right')
result
```

```python
# 57 Page
# how = outer
# full outer join : 합집합
result = pd.merge(left, right, on=['key1','key2'], how='outer')
result
```

```python
# indicator : merge 컬럼 생성 : 어느 데이터쪽에서 왔습니다.
result = pd.merge(left, right, on=['key1','key2'], how='outer', indicator=True)
result
```

#### merge와 join의 결정적 차이 
- pd.merge: 특정 컬럼(열)의 값을 기준으로 합칩니다. (예: 'ID'라는 칸의 값이 같은지 확인)

- df.join: 기본적으로 인덱스(Index, 행 번호/이름)를 기준으로 합칩니다.

### 데이터 결합 (JOIN)

조인은 인덱스를 기준으로 합친다.

### 데이터 변환 (CROSSTAB)

```python
# 81 Page (데이터변환 연습)
# df 데이터 생성
df = pd.DataFrame({'factor1': ['A', 'A', 'A', 'A', 'A','B','B','B', 'B'],
                   'factor2':['a','b','c','a','b','a','b', 'c','c'],
                   'value':[11,13,12,12,14,12,13,11,12]})
df
```

```python
# index, columns 지정
pd.crosstab(df.factor1, df.factor2)
```

행(index)에는 factor1(A, B)을 두어라!

열(columns)에는 factor2(a, b, c)를 두어라!

동작: "A이면서 a인 애가 몇 명이지? B이면서 c인 애는 몇 명이지?"를 전수조사해서 개수를 세라!


factor1과 factor2의 항목들이 서로 교차하는 지점의 데이터 개수(빈도)를 집계  

별도 설정이 없으면 단순히 '건수'만 계산함

```python
# 행/열 명칭 부여 및 margin 추가
pd.crosstab(df.factor1, df.factor2, 
            rownames=['Fact1'],colnames=['Fact2'],
            margins=True)
```

```python
# 82 Page
# index-1개, columns-복수 개 컬럼 지정 가능
pd.crosstab(df.factor1, [df.factor2, df.value])
```

```python
# 각 셀의 빈도 기준 정규화 확률 산출
pd.crosstab(df.factor1, df.factor2, margins=True, 
            normalize=True) 
```

normalize: "개수(Count)"를 "비중(Percentage)"으로 바꿔주는 마법의 옵션
