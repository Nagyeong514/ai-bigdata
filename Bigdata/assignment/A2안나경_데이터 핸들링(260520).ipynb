### 실습 5: df4, df5, df6을 이용해 concat(좌우 결합)을 실행하라

#### STEP ❶ : 데이터 구성하기

```python
# 데이터 구성하기
df4 = pd.DataFrame({'A': ['A0','A1','A2','A3'], 
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3'], }, index=[0, 1, 2, 3])


df5 = pd.DataFrame({'E': ['E0','E1','E2','E3'], 
                   'F':['F0','F1','F2','F3'],
                   'G':['G0','G1','G2','G3'],
                   'H':['H0','H1','H2','H3'] }, index=[0, 1, 2, 3])


df6 = pd.DataFrame({'I': ['I0','I1','I2','I3'], 
                   'J':['J0','J1','J2','J3'],
                   'K':['K0','K1','K2','K3'],
                   'L':['L0','L1','L2','L3'] }, index=[0, 1, 2, 3])
```

#### STEP ❷ concat을 이용한 데이터 결합

```python
result = pd.concat([df4, df5, df6])     #axis = 1 : 가로 결합

result
```

#### STEP ❸ : parameter 변경

```python
# 가로 결합
result = pd.concat([df4, df5, df6], axis = 1)    
result
```

```python
# 새로운 index 생성
result = pd.concat([df4, df5, df6], axis = 1, ignore_index=True)
result
```

### 실습 7:  pivot을 이용하여 GENDER와 AGEGROUP별 WEIGHT의 평균을 아래와 같이 변환하라.

#### STEP ❶ : 데이터 구성하기

```python
FITNESS = pd.read_csv(r"D:\BigData_수업용\2605619_33기 AI·Big Data 실습 파일, 연습 코드\1. Big Data 분석 기초 실습파일\FITNESS.csv", encoding = 'euc-kr')
FITNESS.head(3)
```

#### STEP ❷ : groupby, agg를 이용한 평균값 계산

```python
# 몸무게 평균
df_group = FITNESS.groupby(['GENDER', 'AGEGROUP']).agg({'WEIGHT': 'mean'})

df_group
```

#### STEP ❸ : pivot을 이용한 변환

```python
# 위에서 계산한거 들고와 (df_group)
df_pivot = df_group.pivot_table(index='GENDER', columns='AGEGROUP', values='WEIGHT')

df_pivot
```

```python
# 위에서 계산한거 들고와 (df_group)
df_pivot = df_group.pivot_table(index='AGEGROUP', columns='GENDER', values='WEIGHT')

df_pivot
```
