```python
import pandas as pd, numpy as np, glob

# 설정값
ECO_KW = ("유기농", "무농약", "무항생제")
SNAPSHOT = pd.Timestamp("2021-09-01")
FUTURE_DAYS = 60

# ---------- 1) 로드 & 정제 ----------
def load_clean():
    # 현재 폴더의 모든 product_data_*.csv 파일을 읽어 합침
    files = sorted(glob.glob("product_data_*.csv"))
    if not files:
        print("경고: 불러올 CSV 파일이 없습니다.")
        return pd.DataFrame()
        
    tx = pd.concat([pd.read_csv(f, dtype=str) for f in files], ignore_index=True)
    tx["일자"] = pd.to_datetime(tx["공급일자_"], errors="coerce")
    tx["구매금액"] = pd.to_numeric(tx["구매금액"], errors="coerce")
    tx["구매수량"] = pd.to_numeric(tx["구매수량"], errors="coerce")
    tx["is_return"] = tx["반품_원거래일자"].notna() | (tx["구매금액"] < 0)
    tx["hour"] = pd.to_numeric(tx["구매시각"].str[:2], errors="coerce")
    tx["dow"] = tx["일자"].dt.dayofweek
    tx["is_eco"] = tx["물품명"].fillna("").str.contains("|".join(ECO_KW))
    tx["basket_id"] = tx["회원번호"] + "_" + tx["일자"].dt.strftime("%Y%m%d")
    return tx

def shannon_entropy(s):
    p = s.value_counts(normalize=True)
    return float(-(p * np.log2(p)).sum())

# ---------- 2) 고객 단위 파생 변수 생성 ----------
def build_customer_features(tx, snapshot=SNAPSHOT, future_days=FUTURE_DAYS):
    past = tx[(tx["일자"] < snapshot) & (~tx["is_return"])].copy()
    g = past.groupby("회원번호")

    # RFM
    rfm = g.agg(
        last_buy=("일자", "max"),
        first_buy=("일자", "min"),
        frequency=("basket_id", "nunique"),
        monetary=("구매금액", "sum"),
        avg_basket=("구매금액", "mean"),
        n_lines=("구매금액", "size"),
    )
    rfm["recency_days"] = (snapshot - rfm["last_buy"]).dt.days
    rfm["tenure_days"]  = (rfm["last_buy"] - rfm["first_buy"]).dt.days.clip(lower=1)

    # 재구매 주기
    def gap_stats(s):
        d = np.sort(s.unique())
        if len(d) < 2:
            return pd.Series({"gap_mean": np.nan, "gap_std": np.nan})
        gaps = np.diff(d).astype("timedelta64[D]").astype(int)
        return pd.Series({"gap_mean": gaps.mean(), "gap_std": gaps.std()})
    
    cyc = past.groupby("회원번호")["일자"].apply(gap_stats).unstack()
    cyc["gap_cv"] = (cyc["gap_std"] / cyc["gap_mean"]).replace([np.inf, -np.inf], np.nan)

    # 카테고리/매장/친환경
    div = g["물품대분류"].apply(shannon_entropy).rename("cat_entropy")
    n_cat = g["물품대분류"].nunique().rename("n_categories")
    eco = g["is_eco"].mean().rename("eco_ratio")
    store_focus = g["구매매장"].apply(lambda s: s.value_counts(normalize=True).iloc[0]).rename("store_focus")

    # 최근 추세
    recent = past[past["일자"] >= snapshot - pd.Timedelta(days=60)]
    base   = past[past["일자"] <  snapshot - pd.Timedelta(days=60)]
    rt = recent.groupby("회원번호")["구매금액"].sum().rename("spend_recent60")
    bt = base.groupby("회원번호")["구매금액"].mean().rename("spend_base_mean")

    behav = g.agg(weekend_ratio=("dow", lambda s: (s >= 5).mean()),
                  morning_ratio=("hour", lambda s: (s < 12).mean()))

    ret = tx[tx["일자"] < snapshot].groupby("회원번호")["is_return"].mean().rename("return_rate")

    feat = (rfm.join([cyc, div, n_cat, eco, store_focus, rt, bt, behav, ret]))
    feat["spend_recent_ratio"] = feat["spend_recent60"] / (feat["monetary"] + 1)

    # 회원 속성 (파일이 있는 경우에만 실행)
    try:
        mem = pd.read_csv("member_data.csv", dtype=str)
        mem["연령"] = pd.to_numeric(mem["연령"], errors="coerce")
        mem["is_vip"] = (mem["Gold_member"] == "VIP").astype(int)
        mem["has_delivery"] = (mem["배송서비스신청여부"] == "신청").astype(int)
        mem["has_alarm"] = (mem["모바일알람여부"] == "수신").astype(int)
        mem["gu"] = mem["주소"].str.split("-").str[0]
        feat = feat.join(mem.set_index("회원번호")[["연령", "성별", "is_vip", "has_delivery", "has_alarm", "gu"]])
    except FileNotFoundError:
        print("member_data.csv 파일이 없어 회원 속성 결합을 건너뜁니다.")

    # 라벨 생성
    fut = tx[(tx["일자"] >= snapshot) & (tx["일자"] < snapshot + pd.Timedelta(days=future_days)) & (~tx["is_return"])]
    repurchasers = set(fut["회원번호"].unique())
    feat["y_repurchase"] = feat.index.isin(repurchasers).astype(int)
    
    return feat.reset_index()

# 데이터 로드 및 피처 생성 실행
tx = load_clean()
if not tx.empty:
    feat = build_customer_features(tx)
    print("전처리 완료! 피처 테이블 크기:", feat.shape)
```

### Model Pipline

```python
import lightgbm as lgb
from sklearn.model_selection import GroupKFold
from sklearn.metrics import roc_auc_score

# ===================================================================
# 모델 A. 이탈/재구매 예측
# ===================================================================
def train_churn(feat_df):
    y = feat_df["y_repurchase"]
    # 모델 학습에 불필요하거나 누수가 우려되는 컬럼 제외
    drop_cols = ["회원번호", "last_buy", "first_buy", "y_repurchase", 
                 "spend_recent60", "spend_base_mean"]
    X = feat_df.drop(columns=[c for c in drop_cols if c in feat_df.columns]).copy()
    
    # 범주형 변수 처리
    cat_features = [c for c in ["성별", "gu"] if c in X.columns]
    for c in cat_features:
        X[c] = X[c].astype("category")

    gkf = GroupKFold(n_splits=5)
    aucs = []
    
    print("[모델A] 학습 및 교차 검증 시작...")
    for tr_idx, va_idx in gkf.split(X, y, groups=feat_df["회원번호"]):
        m = lgb.LGBMClassifier(n_estimators=400, learning_rate=0.03, num_leaves=31,
                               subsample=0.8, colsample_bytree=0.8,
                               class_weight="balanced", random_state=42, verbose=-1)
        m.fit(X.iloc[tr_idx], y.iloc[tr_idx], categorical_feature=cat_features)
        prob = m.predict_proba(X.iloc[va_idx])[:, 1]
        aucs.append(roc_auc_score(y.iloc[va_idx], prob))
        
    print(f"-> 평균 AUC = {np.mean(aucs):.4f} ± {np.std(aucs):.4f}")
    return m

# ===================================================================
# 모델 B. 연관 규칙 (장바구니 분석)
# ===================================================================
def train_recommender(tx_df):
    from itertools import combinations
    pos = tx_df[~tx_df["is_return"]].copy()
    baskets = pos.groupby("basket_id")["물품중분류"].apply(lambda s: set(s))
    
    pair = {}
    for items in baskets:
        for a, b in combinations(sorted(items), 2):
            pair[(a, b)] = pair.get((a, b), 0) + 1
            
    n = len(baskets)
    item_counts = pos.groupby("물품중분류")["basket_id"].nunique()
    sup = item_counts / n
    
    rules = []
    for (a, b), count in pair.items():
        support = count / n
        confidence = count / (sup[a] * n)
        lift = confidence / sup[b]
        if support > 0.005 and lift > 1.5:
            rules.append((a, b, support, confidence, lift))
            
    rules_df = pd.DataFrame(rules, columns=["선행", "후행", "support", "confidence", "lift"])
    rules_df = rules_df.sort_values("lift", ascending=False)
    print(f"[모델B] 유효 규칙 수: {len(rules_df)}")
    return rules_df

# ===================================================================
# 모델 C. 수요 예측
# ===================================================================
def train_demand(tx_df):
    d = tx_df[~tx_df["is_return"]].groupby(["일자", "구매매장", "물품중분류"]) \
          .agg(qty=("구매수량", "sum")).reset_index()
    d = d.sort_values("일자")
    
    # Lag 피처 생성
    for lag in (7, 14):
        d[f"lag{lag}"] = d.groupby(["구매매장", "물품중분류"])["qty"].shift(lag)
    d["roll7"] = d.groupby(["구매매장", "물품중분류"])["qty"].shift(1).rolling(7).mean()
    d = d.dropna()

    cut = d["일자"].quantile(0.8)
    tr, va = d[d["일자"] <= cut].copy(), d[d["일자"] > cut].copy()
    
    feats = ["lag7", "lag14", "roll7"]
    cats = ["구매매장", "물품중분류"]
    
    for c in cats:
        tr[c] = tr[c].astype("category")
        va[c] = va[c].astype("category")

    m = lgb.LGBMRegressor(n_estimators=500, learning_rate=0.05, num_leaves=63, random_state=42, verbose=-1)
    m.fit(tr[feats + cats], tr["qty"], categorical_feature=cats)
    
    pred = m.predict(va[feats + cats])
    wmape = np.sum(np.abs(pred - va["qty"])) / np.sum(va["qty"])
    print(f"[모델C] 수요 예측 WMAPE: {wmape:.3f}")
    return m

# 실행
print("=" * 30)
model_churn = train_churn(feat)
print("=" * 30)
rules_recom = train_recommender(tx)
print("=" * 30)
model_demand = train_demand(tx)
```

```python
import pandas as pd, numpy as np, glob

# ── 판정 기준 설정 ───────────────────────────────────────────────────
CERTIFIED   = ["유기농", "무농약", "무항생제"]                        # 인증급
ECO_NATURAL = ["동물복지", "방사", "유정란", "우리밀", "토종",         # 친환경 추정
               "자연", "천연", "무첨가", "무가당", "무방부", "청정",
               "들녘", "비건", "제철"]
FRESH_CATS  = ["채소", "과실", "축산물", "수산"]                      # 신선 카테고리
ORIGIN      = ["국산", "국내산"]                                     # 원산지 참고용
# ───────────────────────────────────────────────────────────────────

def has_keywords(s, kws):
    return s.fillna("").str.contains("|".join(kws), regex=True)

def classify_eco_level(df):
    """우선순위: certified > eco_natural > fresh > other"""
    name = df["물품명"]
    lvl = np.full(len(df), "other", dtype=object)
    
    fresh_mask   = df["물품대분류"].isin(FRESH_CATS)
    eco_mask     = has_keywords(name, ECO_NATURAL)
    cert_mask    = has_keywords(name, CERTIFIED)
    
    lvl[fresh_mask.values] = "fresh"
    lvl[eco_mask.values]   = "eco_natural"
    lvl[cert_mask.values]  = "certified"  # 최우선 순위로 마지막에 덮어씀
    
    df["eco_level"] = lvl
    df["is_origin"] = has_keywords(name, ORIGIN)
    return df

def get_amt_ratio(df, mask):
    total_amt = df["구매금액"].sum()
    return df.loc[mask, "구매금액"].sum() / total_amt if total_amt != 0 else 0

# 1) 데이터 로드 및 등급 분류 (tx가 메모리에 있으면 재사용)
if 'tx' not in locals():
    print("데이터 로딩 중...")
    tx_raw = pd.concat([pd.read_csv(f, dtype=str) for f in sorted(glob.glob("product_data_*.csv"))], ignore_index=True)
    tx_raw["구매금액"] = pd.to_numeric(tx_raw["구매금액"], errors="coerce")
    tx_raw["구매수량"] = pd.to_numeric(tx_raw["구매수량"], errors="coerce")
    tx_raw["일자"] = pd.to_datetime(tx_raw["공급일자_"], errors="coerce")
    tx_raw["is_return"] = tx_raw["반품_원거래일자"].notna() | (tx_raw["구매금액"] < 0)
    tx = tx_raw[~tx_raw["is_return"]].copy()

# 등급 분류 적용
tx = classify_eco_level(tx)
total_amt = tx["구매금액"].sum()

# 2) 결과 출력
print("=== 등급별 판매금액 분해 ===")
eco_summary = tx.groupby("eco_level")["구매금액"].agg(["sum", "size"])
eco_summary["금액비중%"] = (eco_summary["sum"] / total_amt * 100).round(2)
print(eco_summary.reindex(["certified", "eco_natural", "fresh", "other"]).to_string())

print("\n=== 정의별 '친환경/유기농' 금액 비중 ===")
narrow_mask = tx["eco_level"].eq("certified")
mid_mask    = tx["eco_level"].isin(["certified", "eco_natural"])
broad_mask  = tx["eco_level"].isin(["certified", "eco_natural", "fresh"])

print(f"① 좁은 정의(인증만)          : {get_amt_ratio(tx, narrow_mask)*100:5.2f} %")
print(f"② 중간 정의(+친환경추정)      : {get_amt_ratio(tx, mid_mask)*100:5.2f} %")
print(f"③ 넓은 정의(+신선식품 전체)   : {get_amt_ratio(tx, broad_mask)*100:5.2f} %")

print("\n=== 대분류별 '친환경(중간정의)' 금액 비중 ===")
tx["is_eco_mid"] = tx["eco_level"].isin(["certified", "eco_natural"])
cat_summary = tx.groupby("물품대분류").apply(
    lambda d: pd.Series({
        "친환경금액": d.loc[d["is_eco_mid"], "구매금액"].sum(),
        "총금액": d["구매금액"].sum()
    }), include_groups=False
)
cat_summary["비중%"] = (cat_summary["친환경금액"] / cat_summary["총금액"] * 100).round(2)
print(cat_summary.sort_values("비중%", ascending=False)[["비중%"]].head(15).to_string())

print("\n=== 고객 관점 (중간 정의 기준) ===")
cust_stats = tx.groupby("회원번호").agg(
    has_bought=("is_eco_mid", "max"),
    personal_ratio=("is_eco_mid", "mean")
)
print(f"친환경 1회 이상 구매 고객 비율 : {cust_stats['has_bought'].mean()*100:.2f} %")
print(f"고객 1인당 평균 친환경 비중   : {cust_stats['personal_ratio'].mean()*100:.2f} %")
```
