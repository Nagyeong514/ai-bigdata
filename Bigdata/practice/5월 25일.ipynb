## 선형 회귀

```python
# 28 Page
# 선형 회귀 분석 : formula(y ~ X1 + X2 + ...) 이용. 상수항(Intercept)은 자동으로 추가됨
reg_model = smf.ols(formula = "MINUTES ~ UNITS", data = df_raw)
reg_result = reg_model.fit()        # fit 학습
print(reg_result.summary())         # summary : 결과 반환
```

모델이 OLS  
Observer :   
resi  
r-squared : 0.987( 0.9뭐를 제곱해봐래)  
p 값 : e의 -2  
ALC가 적으면적을수록 좋다.  
intercept    아까 계산해라 했던거 codf => std err    
jarque bera : 이거도 0.05 보다 크다 정규성  


2. 두 번째 이미지: 회귀 모델 생성 및 결과 해석 (OLS)

드디어 수식을 만든 단계입니다. MINUTES = 15.5088 * UNITS + 4.1617이라는 식을 찾았네요.

    R-squared (0.987): 이 모델의 설명력이 98.7%라는 뜻입니다. 아주 훌륭한 모델입니다.

    P>|t| (0.000): UNITS 변수의 p-값이 0.05보다 작으므로, 이 변수는 결과에 영향을 주는 아주 유의미한 변수입니다.

    Jarque-Bera: 잔차의 정규성을 확인하는 지표인데, 메모에 적으신 대로 0.05보다 크면 "잔차가 정규분포를 따른다"고 보고 모델을 신뢰할 수 있습니다.

    AIC / BIC: 모델의 복잡도를 평가하는 지표로, 낮을수록 좋은 모델입니다.

```python
# 29 Page
# 잔차 산점도: X-예측값, Y-잔차 
plt.scatter(reg_result.fittedvalues, reg_result.resid)
plt.xlabel("fitted value"); plt.ylabel("residual")
plt.axhline(y = 0, linestyle = "--")
```

0을 중심으로 랜덤하게 존재 한다 : 등분산서이 있다.
커지는 형태로 : 이분산성 : 문제가 잇는 모델

3. 세 번째 이미지: 잔차 분석 (Residual Plot)

모델이 못 맞힌 부분인 '잔차'를 시각화해서 모델의 가정을 검사하는 단계입니다.

    잔차 산점도: 예측값(fitted value)에 따라 잔차가 특정 패턴 없이 0을 중심으로 무작위로 흩어져 있어야 합니다.

    해석: 만약 잔차가 휘어지거나 특정 모양을 그리면 "모델이 데이터의 특성을 다 잡아내지 못했구나"라고 판단하는데, 지금 그래프는 골고루 잘 퍼져 있는 편이라 모델이 적절하다고 볼 수 있습니다.

```python
# 잔차 정규성 검정:Q-Q plot. s-standardized line, 
#  r-regression line, q-quartiles
qqplot(reg_result.resid, line = "q")
plt.show()              # 하나만 쇼
```

정규분포 직선 (적색 선)을 중심으로 분포하므로 정규 분포를 따름

```python
# 30 Page
# 잔차의 독립성 확인
plt.scatter(df_raw.index, reg_result.resid, marker="o") # y축은 잔차
plt.xlabel("Data") ; plt.ylabel("Residual") #인덱스

plt.axhline(y=0,linestyle="-", label='CL')  # 스타일: 실선, 평균값
plt.axhline(y=3*np.std(reg_result.resid),linestyle="--",color="red",label='UCL' )   #3시그마를 반영함, 스타일
plt.axhline(y=-3*np.std(reg_result.resid),linestyle="dashed",color="red",label='LCL')   # 마이너스 3시그마 (기본값은 파란색 선)
plt.legend() ;      #범례를 찍고
plt.show()
```
