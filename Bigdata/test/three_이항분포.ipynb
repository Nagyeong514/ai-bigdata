도장공정에서 광택도 불량이 40%나 된다고 함  
▪ 3대의 차량을 임의로 선택했을 때 불량대수가 각각 0, 1, 2, 3 대가 나올 확률은
얼마인가요?

p = 0.4  
3대니까 n = 3

```python
# 패키지 불러오기 (기본)

# 데이터 구성 패키지 : Series, DataFrame
import pandas as pd

# 행렬 연산 패키지
import numpy as np

# 통계 분석 패키지
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest

# 그래프 표시
%matplotlib inline
```

```python
n = 3
for i in range(n+1):
    prob = stats.binom.pmf(k = i, n = n, p = 0.4)
    print("P(X={0}) = {1:.3f}".format(i,prob))
```

▪ 도장공정에서 광택도 불량이 30%나 된다고 함

```python
n = 5
for i in range(n+1):
    prob = stats.binom.pmf(k = i, n = n, p = 0.3)
    print("P(X={0}대 불량일 확률.) = {1:.3f}".format(i,prob))
```

# 제미나이 문제

📝 [손코딩 연습 문제 - 중상 난이도]

문제:

어느 반도체 공정의 불량률은 10%($p=0.1$)라고 합니다. 이 공정에서 생산된 제품 중 10대($n=10$)를 임의로 추출했을 때, 다음 요구사항에 맞는 파이썬 코드를 작성하시오.

누적 확률 계산: 불량대수가 3대 이하(0, 1, 2, 3대)일 확률을 stats.binom.cdf 함수를 사용하여 구하고 출력하시오.

반복문 활용: 불량대수가 2대 이상 5대 이하일 확률을 구하기 위해, for문과 stats.binom.pmf를 사용하여 각 확률을 더한 뒤 최종 결과를 출력하시오. (결과 변수명: prob_range)

마무리: 모든 결과값은 소수점 넷째 자리까지 반올림하여 표시하시오.

```python
# 누적 확률 계산
stats.binom.cdf(k=3, n=10, p=0.1)
```

```python
# for문과 stats.binom.pmf를 사용하여 각 확률을 더한 뒤 최종 결과를 출력하시오.
# 아 그니까 pmf로 누적을 구하려니까;;;;

n = 10
prob_range = 0

for i in range(2,6):
    prob = stats.binom.pmf(k = i, n = n, p = 0.1)
    prob_range += prob
    print("P(X={0}) = {1:.4f}".format(i,prob))

print("최종 합계(2~5대) = {0:.4f}".format(prob_range))
```

```python
# 위를 cdf로 하면 쉽지
prob_range = stats.binom.cdf(5, 10, 0.1) - stats.binom.cdf(1, 10, 0.1)
print("{0:.4f}".format(prob_range))
```
