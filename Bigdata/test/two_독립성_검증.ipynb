# 실습 문제

노트북 컴퓨터의 6가지 제품(A, B, C, D, E, F)에 대하여 7가지 제품의 이미지를 중복 선택  
할 수 있도록 32 명의 소비자를 대상으로 설문 조사한 Table이다. 각 제품에 대해 이미지에 대해 차이가 있겠는가? (유의수준 0.05)

```python
# df 생성
df = pd.DataFrame({'A':[18, 1, 8, 7, 10, 9, 10], 'B':[8, 2, 14, 5, 5, 9, 4],
                   'C':[4, 1, 3, 4, 9, 5, 4], 'D':[4, 1, 2, 3, 2, 7, 3], 'E':[3, 1, 3, 1, 1, 1, 1], 'F':[3, 25, 8, 10, 2, 1, 7]}, index=['A/S', '가격', '성능', '확장성', '디자인', '안정성', '기능성'])

df
```

```python
# chi-square test 실행
chi, pval, dof, expected = stats.chi2_contingency(df)
chi, pval, dof, expected
```

```python
# chi-square test 결과 출력
print( "chi-square test ")
print( ' chisq: {0:0.2f}' .format(chi))
print( ' p-value : {0:0.3f}'.format(pval))
print( ' degree of freedom: {}' .format(dof))
print( ' expected value: \n{}' .format(expected.round(1)))
```

엥...? 실측값이 chisq : 98.27 인데 기대치가 11. 8. 뭐 이렇네요 간극이 엄청 큰데요..? 이건 귀무 가설을 기각 시켜야겠습니다..   
p 값도 0 퍼 입니다.. 대립 가설을 채택한다고 했을떄 오류가 날 가능성이 없습니다. 야호 제 주장이 맞았네요 그럼 대립 가설을 채택하도록 하겠습니다.

📝 [실습 문제] 카이제곱 검정: 연령대별 음료 선호도 분석
문제:
어느 카페에서 연령대(10대, 20대, 30대)에 따라 선호하는 음료 메뉴(커피, 에이드, 차)의 차이가 있는지 알아보기 위해 300명을 대상으로 설문 조사를 실시하여 아래와 같은 교차표(Table)를 얻었습니다.

연령대에 따라 선호하는 음료 메뉴에 차이가 있는지 유의수준 0.05에서 검정하시오.

```python
# df 생성
df = pd.DataFrame({'커피':[20, 70, 60], '에이드':[50, 30, 10],
                   '차':[10, 20, 30]}, index=['10대', '20대', '30대'])

df
```

```python
# chi-square test 실행
chi, pval, dof, expected = stats.chi2_contingency(df)
chi, pval, dof, expected
```

```python
# chi-square test 결과 출력
print( "chi-square test ")
print( ' chisq: {0:0.2f}' .format(chi))
print( ' p-value : {0:0.3f}'.format(pval))
print( ' degree of freedom: {}' .format(dof))
print( ' expected value: \n{}' .format(expected.round(1)))
```
