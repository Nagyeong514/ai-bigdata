### Q. 부품수리시간과 부품 수간에 관계를 분석하기 위한 상관분석을 하세요.

```python
#  직접 입력 
import pandas as pd
import numpy as np
from scipy import stats

df1 = [1, 2, 3, 4, 4, 5, 6, 6, 7, 8]
df2 = [23, 29, 49, 64, 74, 87, 96, 97, 109, 119]
```

```python
import matplotlib.pyplot as plt

# 산점도 그리기 (가장 먼저 수행)
plt.scatter(df1, df2)
plt.xlabel("Repair Time")
plt.ylabel("Parts Count")
plt.show()
```

```python
# 두 변수 각각 정규성 확인
_, p_norm1 = stats.shapiro(df1)
_, p_norm2 = stats.shapiro(df2)
print(f"정규성 p-value: {p_norm1:.4f}, {p_norm2:.4f}")
```

귀무가설($H_0$): 수리 시간과 부품 수 사이에는 관계가 없다. ($r = 0$)  
대립가설($H_1$): 수리 시간과 부품 수 사이에는 관계가 있다. ($r \neq 0$)

```python
corr, pval = stats.pearsonr(df1, df2)
```

```python
print("Correlation Analysis")
print('corr: {0:0.3f}'.format(corr))
print( 'p-value : {0:0.3f}'.format(pval))
```

통계적 판정: "p-value가 0.05보다 작으므로 귀무가설을 기각하고 대립가설을 채택한다.  

"상관계수 해석: "상관계수 $r=0.989$이므로, 두 변수 사이에는 통계적으로 유의미한 매우 강한 양의 상관관계가 존재한다."

0이네요 대립가설 채택 : 부품수리 시간과 부품수 간에 매우 높은 상관성이 존재한다.
