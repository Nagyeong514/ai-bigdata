# 베스트셀러 크롤링 순서

---

## 1. 준비
```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```
- `requests` : 웹페이지 HTML 가져오기
- `BeautifulSoup` : HTML 파싱 (원하는 데이터 뽑기)
- `pandas` : 데이터프레임 생성 & CSV 저장

---

## 2. 페이지 HTML 가져오기
```python
url = "크롤링할 사이트 URL"
h = {"User-Agent": "Mozilla/5.0"}
res = requests.get(url, headers=h)
soup = BeautifulSoup(res.text, "html.parser")
```
- `User-Agent` 없으면 봇으로 인식해서 차단당할 수 있음
- F12 → Network 탭에서 URL 확인

---

## 3. 원하는 태그 찾기 (CSS 셀렉터)
```python
# F12 → 원하는 요소 우클릭 → Copy → Copy selector
sel = "ul.list_books > li"
items = soup.select(sel)   # 리스트로 반환
```
- `select()` : 여러 개 → 리스트
- `select_one()` : 하나만

---

## 4. 각 항목에서 데이터 뽑기
```python
data = []
for item in items:
    title = item.select_one(".title").text.strip()
    author = item.select_one(".author").text.strip()
    data.append({"제목": title, "저자": author})
```
- `.text` : 태그 안 텍스트 추출
- `.strip()` : 앞뒤 공백 제거

---

## 5. 데이터프레임 생성 & CSV 저장
```python
df = pd.DataFrame(data)
df.to_csv("bestseller.csv", index=False, encoding="utf-8-sig")
df
```
- `utf-8-sig` : 한글 깨짐 방지
- `index=False` : 인덱스 번호 제외하고 저장

# 교보문고 베스트셀러 크롤링

## 각 단계에서 필요한 자료 찾는 법

| 필요한 것 | 어디서 찾나 |
|---|---|
| **URL** | 브라우저 주소창 또는 F12 → Network 탭 → Request URL |
| **API URL** | F12 → Network 탭 → Fetch/XHR 필터 → 요청 클릭 → General > Request URL |
| **요청 헤더** (User-Agent, API Key 등) | F12 → Network 탭 → 요청 클릭 → Headers 탭 → Request Headers |
| **파라미터** (page, cid 등) | F12 → Network 탭 → 요청 클릭 → Payload 탭 |
| **CSS 셀렉터** | F12 → 원하는 요소 우클릭 → 검사 → 태그 우클릭 → Copy → Copy selector |
| **JSON 필드명** | F12 → Network 탭 → 요청 클릭 → Preview 탭에서 구조 확인 |
| **JS 렌더링 여부** | `print("찾는텍스트" in res.text)` → False면 JS 렌더링 |

````python
# 2. 페이지별 데이터 수집 (1~3페이지)
data = []

for page in range(1, 4):
    p = {"page": page, "per": 20, "period": "003", "bsslBksClstCode": "A"}
    res = requests.get(url, params=p, headers=h)
    books = res.json()["data"]["bestSeller"]

    print(f"\n=== {page}페이지 ===")
    for book in books:
        print(f"{book['prstRnkn']}위: {book['cmdtName']} / {book['chrcName']}")
        data.append({
            "순위": book["prstRnkn"],
            "제목": book["cmdtName"],
            "저자": book["chrcName"],
            "출판사": book["pbcmName"],
            "정가": book["price"],
            "판매가": book["sapr"],
            "할인율": book["dscnRate"],
        })
````
