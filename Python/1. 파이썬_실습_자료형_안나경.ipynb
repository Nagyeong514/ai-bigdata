```python
# 정수형 (Integer)
a = 42        
a = -7        
a = 0
print(type(a))   # <class 'int'>

# 실수형 (Floating-point / Float)
a = 3.14      
a = -0.5
print(type(a))   # <class 'float'>
```

```python
a = 6
b = 4

# 사칙 연산
print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
# 제곱 연산자
print(2 ** 5)   # 32
print(5 ** 2)   # 25

# 나머지 연산자 
print(5 % 2)    # 1
print(4 % 2)    # 0

# 몫 연산자
print(7 // 2)   # 3
print(5 // 3)   # 1
```

```python
# 문자열 포매팅
# 고정된 문자열
print("I eat %d apples." % 3)

# 정수 대입
print("I have %d notebooks." % 5)

# 실수 대입
print("Interest rate is %f" % 3.14)

# 문자열 대입
print("I love %s." % "Spring Boot")

# 숫자 값을 나타내는 변수로 대입
hours = 5
print("I studied for %d hours." % hours)

# 2개 이상의 포맷 코드 함께 사용
print("Name: %s, Age: %d, GPA: %.2f" %("안나경", 24, 4.11))
```

```python
# 문자열 포매팅 출력 형식 제어
# 전체 폭 10, 오른쪽 정렬
print("%10s" % "hi")

# 전체 폭 10, 왼쪽 정렬
print("%-10sjane" % "hi")

# 소수점 자리수 지정
rate = 87.456 
print("%.2f" % rate)

# 퍼센트(%) 함께 출력
print("%d%%" % 95)

# 전체 폭 12, 소수점 이하 3자리
print("%12.3f" % rate)
```

```python
# format() 함수 사용
# 순서 기반 포매팅
print("I eat {} apples".format(3))

# 인덱스 기반 포매팅
print("{0} loves {1}! {0} loves {2} too".format("Java", "Python", "C++"))

# 이름 기반 포매팅 (숫자는 따옴표 없이 사용가능)
print("I ate {number} apples. so I was sick for {day} days.".format(number=10, day=3))

# 인덱스 + 이름 혼용
print("I ate {0} apples. so I was sick for {day} days.".format(10, day=3))

# 문자열 리터럴을 직접 넣는 경우 (반드시 따옴표 사용)
print("반값습니다. {name}님!".format(name="안나경"))

# 중괄호{} 문자 출력
print("{{ }}".format())
```

```python
# 오른쪽 정렬 (폭 10)
print("{:>10}".format("hi"))

# 왼쪽 정렬 (폭 10)
print("{:<10}".format("hi"))

# 소수점 자리수 지정
print("{:.2f}".format(rate))

# 퍼센트 출력
print("{:%}".format(0.8745))

# 퍼센트 출력 (%% 불필요)
print()
```

```python
# 가운데 정렬
print("{:^10}".format("hi"))

# 문자로 채우기
print("{:=^10}".format("hi"))
print("{:!<10}".format("hi"))
```
