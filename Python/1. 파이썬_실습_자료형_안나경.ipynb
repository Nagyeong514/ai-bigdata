3. 리스트 복사

```python
# 참조복사(b=a) : a, b 주소값 같음    = 이름만 두 개일 뿐, 메모리상의 실제 몸체(주소)는 하나입니다.
a = [1, 2, 3]
b = a
a[1] = 4


print(a)
print(id(a))

print(b)
print(id(b))
```

```python
# [:] 이용
# 얕은 복사 (b = a[:] 또는 b = a.copy())   =  개념: 원본의 내용을 그대로 베껴서 별도의 새로운 몸체(주소)를 만듭니다.
a = [1, 2, 3]
b =  a[:]
a[1] = 4

print(a)
print(id(a))

print(b)
print(id(b))
```

```python
# copy 모듈 이용
a = [1, 2, 3]
b = a.copy()

print(a)
print(id(a))

print(b)
print(id(b))
```

```python
# shallow copy: 겉껍데기(바깥 리스트)만 새로 만들고, 그 안의 내부 리스트는 주소값만 복사해옵니다.
a = [[1, 2], [3, 4]]
b = a.copy()

a[0][0] = 99

print(a)
print(id(a))
# [[99, 2], [3, 4]]

print(b)
print(id(b))
# [[99, 2], [3, 4]]
```

```python
# deep copy : 겉껍데기뿐만 아니라 그 안에 들어있는 모든 객체까지 통째로 새로 만듭니다.
import copy     # 파이썬 기본 내장 함수

a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)

a[0][0] = 99



print(a)
print(id(a))
# [[99, 2], [3, 4]]

print(b)
print(id(b))
# [[1, 2], [3, 4]]
```
