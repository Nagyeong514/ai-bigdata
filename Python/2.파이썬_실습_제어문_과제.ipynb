과제 1. 1부터 10까지의 숫자 중에서 홀수만 출력하되, while문과 continue를 반드시 사용하시오.

```python
a = 0

while a < 10:
    a += 1
    if a%2 == 0:
        continue

    print(a)
```

과제 3. 각 학생의 점수를 검사하여 점수가 60점 이상이면 “합격”, 그렇지 않으면 “불합격”을 출력하시오. 또한, 합격한 학생 이름만 모은 리스트와 합격/불합격 인원수를 마지막에 출력하시오.

** 출력 예시 **
<pre style="font-size:14px;">
철수 : 합격
영희 : 불합격
민수 : 합격
지은 : 불합격
합격자: ['철수', '민수']
합격: 2명, 불합격: 2명
</pre>

```python
students = [("철수", 85), ("영희", 42), ("민수", 73), ("지은", 58)]

pass_list = []
pass_count = 0
fail_count = 0

number = 0
for name, score in students:
    number += 1
    if score >= 60:
        print(f"{name} : 합격")
        pass_count += 1
        pass_list.append(name)
    else: 
        print(f"{name} : 불합격")
        fail_count += 1

print(f"합격자: {pass_list}")
print(f"합격: {pass_count}명, 불합격: {fail_count}명")



```

과제 4. for와 range 함수를 사용하여 2단부터 9단까지의 구구단을 출력하시오. 각 단은 한 줄에 출력하되, 숫자 사이는 공백으로 구분하시오.

** 출력 예시 **
<pre style="font-size:14px;">
2 4 6 8 10 12 14 16 18 
3 6 9 12 15 18 21 24 27 
4 8 12 16 20 24 28 32 36 
5 10 15 20 25 30 35 40 45 
6 12 18 24 30 36 42 48 54 
7 14 21 28 35 42 49 56 63 
8 16 24 32 40 48 56 64 72 
9 18 27 36 45 54 63 72 81 
</pre>

```python
i = 0

for i in range (2, 10):      #2단에서 9단이지
    # 2일떄
    for j in range (1, 10):  # 2 1 = 1 , 2 2 = 4 .. 2 9 = 18
        print(i*j, end=" ")
        
    print()
```
