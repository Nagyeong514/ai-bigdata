### 과제 1
- 몇 개의 재료가 들어갈지 모름.(매개변수명: ingredients)
- 만약 입력된 재료가 하나도 없다면, '주문할 재료를 선택해주세요.'라고 출력하고, **즉시 함수를 종료하기**

```python
# 과제 1번
def order_snack(*ingredients):
    # 1. 재료가 없는 경우(강제 종료) return
    if len(ingredients) == 0:
        print("주문할 재료를 선택해주세요.")
        return
    
    else:
        print(f"주문하신 재료: {ingredients}")
        print("주문이 완료됨")
    
     
        
```

```python
order_snack()
```

```python
order_snack("치즈", "햄", "토마토")
```

### 과제 2

> halo.txt 파일 내용을 한 줄씩 읽어 모두 대문자로 바꿔주는 프로그램을 작성하시오.
- 파일명 입력 : input() 함수를 사용하여 읽고 싶은 파일 이름을 입력받기
- 파일 읽기 : with문과 readlines() 함수를 사용하여 파일 내용을 읽기
- 대문자 변환 : upper() 함수를 사용하기
- 파일 인코딩 형식은 'utf-8'로 사용할 것


```python
#Python 파일 : sys2.py

import sys
args = sys.argv[1:]

ori_file = args[0]
cp_file = ori_file + "_cp.txt"


with open(ori_file, 'r', encoding="utf-8" ) as f:
    content = f.read()

with open(cp_file, 'w', encoding="utf-8" ) as f:
    f.write(content)
    
```

### 과제 4

- 여러 개의 숫자를 입력받음 = 입력받는 숫자의 개수는 고정하지 않음.
- 입력받은 숫자의 제곱 값을 계산함.
- 계산된 결과를 리스트 형태로 반환하는 get_even_squares() 함수를 작성하시오.

```python
def get_even_squares(*numbers):
    result=[]

    for num in numbers:
        if num % 2 == 0:
            result.append(num ** 2)
    
    return result
    

print(get_even_squares(1, 2, 3, 4, 5, 6))
print(get_even_squares(10, 20, 30))
```
