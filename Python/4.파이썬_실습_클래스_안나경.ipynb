## 예제 4 -p20

##### - 최소 잔액을 유지해야 하는 계좌 클래스를 만드세요
###### -기본적으로 기존 실습 문제1의 계좌와 기능 동일함
###### -하지만 출금 시 잔액이 최소 잔액 미만이 되면 출금 못함 그리고 “최소 잔액을 유지해야 합니다” 메시지 출력
###### -힌트: 상속 및 오버라이딩 활용

```python
class MinimumBalanceAccount(BankAccount):       # class <현재 클래스 이름>(<상속할 클래스 이름>)
    def __init__(self, balance=0, name="none", min_bal=0):
        # 채우기
        # super : super() 함수는 자식 클래스에서 부모 클래스(또는 상위 클래스)의 메서드나 속성에 접근하기 위해 사용하는 함수
        super().__init__(balance, name) #다른 것도 있었으니까 우리는 이거를 상속 받아와서 쓸거다 ~
        self.min_bal = min_bal
        

    def withdraw(self, amount):
        if self.balance - amount < self.min_bal:
            print("최소 잔액을 유지해야합니다.")
    
        else:
            super().withdraw(amount)
```

```python
acc = MinimumBalanceAccount(1000, "Kim", 500)
```

```python
acc.withdraw(100)
print(acc.balance)
```

```python
acc.withdraw(600)
print(acc.balance)
```

## 예제8 -p40

##### - 비밀번호 검사 (pass 사용)
###### - 비밀번호는 문자열 리스트에 저장되어 있고 조건에 맞는 비밀번호를 출력
###### - 6글자 미만이거나 숫자로만 된 비번은 예외 발생 (아무 메시지 없이 넘어갈 것)

```python
passwords = ["1234", "abcd12", "pass", "hello99", "999999", "Python3"]

for pw in passwords:
    # 채우기]
    try:
        if len(pw) < 6:
            raise Exception("길이 부족")

        if pw.isdigit():
            raise Exception("숫자만 존재")
        
        print(pw)

    except Exception as e :
        print(pw, ':', e)
        pass
```
