```python
import copy

list1 = [1, 2, [3, 4]]  # 내부에 리스트가 또 있는 가변 객체
list_shallow = copy.copy(list1) # 얕은 복사 수행

print(id(list1) == id(list_shallow))      # False (껍데기는 새로 만듦)
print(id(list1[2]) == id(list_shallow[2])) # True (내부 리스트 주소는 공유함)
```

```python
list_deep = copy.deepcopy(list1) # 깊은 복사 수행

print(id(list1) == id(list_deep))      # False (껍데기 새로 만듦)
print(id(list1[2]) == id(list_deep[2])) # False (내부 리스트까지 새로 만듦)
```

함수가 정의(Define)되는 시점에 []라는 리스트 객체가 메모리에 한 번 생성됩니다.

```python
def add_item(item, items = None):
    if items is None:
        items = [] 
    items.append(item)
    return items
print(add_item(1))
print(add_item(2))
```
