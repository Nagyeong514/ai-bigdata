## 1-1. Immutable, Mutable

##### Python 인터닝 범위 : -5 ~ 256
##### JAVA 인터닝 범위 : -128 ~ 127

```python
# 불변 객체 (문자열)
s1 = "hello"
s2 = "hello"
print(f"s1 주소: {id(s1)}")
print(f"s2 주소: {id(s2)}")
print(f"s1과 s2의 값 비교 (==): {s1 == s2}") # True (글자가 같음)
print(f"s1과 s2의 주소 비교 (is): {s1 is s2}") # True (인터닝으로 주소도 같음)

print("-" * 30)

# 가변 객체 (리스트)
list1 = [1, 2, 3]
list2 = [1, 2, 3]
print(f"list1 주소: {id(list1)}")
print(f"list2 주소: {id(list2)}")
print(f"리스트 내용 비교 (==): {list1 == list2}") # True (값은 같음)
print(f"리스트 주소 비교 (is): {list1 is list2}") # False (메모리 위치는 다름)

print("-" * 30)

#  튜플 생성 및 주소 확인
t1 = (1, 2, 3)
t2 = (1, 2, 3)

print(f"t1 주소: {id(t1)}")
print(f"t2 주소: {id(t2)}")
# 튜플은 값이 같더라도 리터럴 최적화가 이루어지지 않으면 보통 주소가 다릅니다.
print(f"튜플 내용 비교 (==): {t1  == t2}") # True (값은 같음)
print(f"튜플 주소 비교 (is): {t1  is t2}") # False (메모리 위치는 다름)


#id() 함수: 메모리 주소값 반환
```

```python
# 1. 작은 정수 테스트 (-5 ~ 256 범위 안)
# 파이썬이 미리 메모리에 만들어 두고 재사용하는 구간입니다.
n1 = 256
n2 = 256
print(f"n1 주소: {id(n1)}")
print(f"n2 주소: {id(n2)}")
print(f"작은 숫자(-5 ~ 256) 주소 비교 (is): {n1 is n2}") # True

print("-" * 30)

# 2. 큰 정수 테스트 (범위 밖)
# 이 범위를 벗어나면 원칙적으로는 매번 새로운 객체를 생성합니다.
n3 = 1000
n4 = 1000
print(f"n3 주소: {id(n3)}")
print(f"n4 주소: {id(n4)}")
print(f"큰 숫자 주소 비교 (is): {n3 is n4}") # False (또는 최적화에 의해 True)

print("-" * 30)

# 2. 인터닝 범위를 벗어난 숫자 (-8)
n5 = -8
n6 = -8
print(f"-8의 주소 : {id(n5)}")
print(f"-8의 주소 : {id(n6)}")
print(f"인터닝 범위를 벗어난 숫자 비교 (is): {n3 is n4}") # False (또는 최적화에 의해 True)


#f: f-string 시작
#{id(n5)}: id(n5) 함수를 실행값 반환
```

## 1-2 형변환 (복습아니긴해)

```python
print(10 / 5)   # 2.0 (암시적 형변환: 나눗셈은 항상 실수형 반환)
print(10 // 5)  # 2 (몫 연산은 정수형 반환)
print(7 + 3.0)  # 10.0 (암시적 형변환: 정수와 실수의 연산은 실수형 반환)

# 파이썬 불리언 선언: 타입을 명시하지 않고 첫 글자를 대문자로 작성
a = True
print(a + 1)    # 2 (True는 산술 연산 시 1로 자동 형변환됨)
```
