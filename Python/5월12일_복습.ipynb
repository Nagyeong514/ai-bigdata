## 1-1. 조회(반환값 있음)와 명령(원본 상태 변화)

```python
# 1. 값을 알려주는 함수 (결과를 보려면 print 필요)
a = [10, 20, 30]
print(a.index(20))  # 20이 어디 있는지 '값'을 돌려줌 -> 출력: 1
print(a)            # 원본은 그대로 [10, 20, 30]


# 2. 직접 바꾸는 함수 (실행 후 원본을 확인)
b = [3, 1, 2]
b.sort()            # b 자체를 [1, 2, 3]으로 바꿔버림 (반환값 없음)
print(b)            # 바뀐 결과 확인 -> 출력: [1, 2, 3]
```

index(): 리스트는 가만히 있고, 궁금한 정보만 리턴합니다. 그래서 print()가 필요합니다.

sort(): 리스트 자체를 수정(Update)합니다. 리턴값이 없어서 print(b.sort())를 하면 None이 나옵니다.

## 1-2 set 함수

```python
# 정의하기
s = set()
s2 = {'d','c','b', 'a'}


# 1. 낱개(덩어리) 추가
s.add('Java')        # {'Java'}
print(s)

# 2. 묶음(여러 덩어리) 추가 (리스트 활용)
s.update(['Python', 'Spring']) # {'Java', 'Python', 'Spring'}
print(s)

# 3. 문자열 쪼개기 주의
s.update('SQL')      # {'Java', 'Python', 'Spring', 'S', 'Q', 'L'}
print(s)

# 4. 숫자 추가 (리스트 필수)
s.update([100, 200]) # {'Java', ..., 100, 200}
print(s)



## 주의
s2.update('안안안되나요/')          #{'나', '되', '안', '얘는', '요'}
#s2.update(123)                   #"보따리를 풀어서 그 안의 내용물을 다 넣어라"라는 명령인데 int(숫자)는 풀수가 없어!
#add(1, 2)                      # 인자를 2개 이상 넣으면 에러가 발생합니다. (무조건 1개만 가능)
```

## 1-3 참조 복사 vs 얕은 복사 vs 깊은 복사

```python
import copy

# 1. 참조 복사 (주소 공유)
a = [1, 2]
b = a
print(f"1. 참조 복사")
print(f"주소: a={id(a)}, b={id(b)}")
print(f"주소 동일 여부: {id(a) == id(b)}") # True
print("-" * 30)

# 2. 얕은 복사 (바깥만 복사)
# shallow copy: 겉껍데기(바깥 리스트)만 새로 만들고, 그 안의 내부 리스트는 주소값만 복사해옵니다.
a = [[1, 2], 3]
b = a.copy()
print(f"2. 얕은 복사")
print(f"바깥 주소 동일 여부: {id(a) == id(b)}") # False
print(f"내부 리스트 주소 동일 여부: {id(a[0]) == id(b[0])}") # True (내부는 공유)
print("-" * 30)

# 3. 깊은 복사 (내부까지 전부 복사)
# deep copy : 겉껍데기뿐만 아니라 그 안에 들어있는 모든 객체까지 통째로 새로 만듭니다.
a = [[1, 2], 3]
b = copy.deepcopy(a)
print(f"3. 깊은 복사")
print(f"바깥 주소 동일 여부: {id(a) == id(b)}") # False
print(f"내부 리스트 주소 동일 여부: {id(a[0]) == id(b[0])}") # False (내부도 새로 생성)
```
