```python
# 1. 함수 정의 (Function)
# 특정 점수를 받아서 등급(A, B, C)을 반환하는 함수입니다.
def check_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    else:
        return "C"

# 2. 리스트와 for 반복문 (For Loop)
# 여러 명의 점수 데이터를 처리합니다.
score_list = [95, 82, 77, 65, 91]
print("--- 전체 성적 현황 ---")

for s in score_list:
    grade = check_grade(s)
    # 80점 이상인 경우에만 우수 문구를 출력하는 if문
    if s >= 80:
        print(f"점수: {s}점 / 등급: {grade} (우수)")
    else:
        print(f"점수: {s}점 / 등급: {grade}")

# 3. while 반복문과 조건문 (While Loop & Input)
# 사용자가 종료를 원할 때까지 계속 점수를 입력받아 등급을 알려줍니다.
print("\n--- 실시간 등급 조회 (종료하려면 'q' 입력) ---")

while True:
    user_input = input("조회할 점수를 입력하세요: ")
    
    # 종료 조건 (if & break)
    if user_input.lower() == 'q':
        print("조회를 종료합니다.")
        break
    
    # 입력값이 숫자인지 확인 후 처리
    score_num = int(user_input)
    result = check_grade(score_num)
    print(f"입력하신 {score_num}점은 {result}등급입니다.")
```
