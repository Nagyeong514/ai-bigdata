### 2-1. if문

- if문의 기본 구조

```python
if 조건A:
    조건A가 참일 때 수행
    ...

    조건B가 참일 때 수행
    ...
elif 조건C:
    조건C가 참일 때 수행
    ...

    ...

```

### 2-3. for문

- for문의 기본 구조

```python
for 변수 in 리스트(또는 튜플, 문자열):
    반복 수행할 문장1
    반복 수행할 문장2
    ...

```

````python
steps = ["first", "second", "third"]

for step in steps:
    print(step)
````

````python
pairs = [(1,2), (2,4), (3,6)]

for a, b in pairs:
    print(a + b)
````

````python
# zip 함수
days = ['월', '화', '수']
menus = ['김밥', '라면', '돈까스']

for day, menu in zip(days, menus):
    print(f"{day}요일 점심: {menu}")
````
